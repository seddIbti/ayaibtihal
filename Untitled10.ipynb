```python

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Charger les images
img1 = cv2.imread('/content/DJI_0020.JPG', 0)  # Remplacez par le chemin de votre image
img2 = cv2.imread('/content/DJI_0026.JPG', 0)  # Remplacez par le chemin de votre image

# Vérifier si les images ont été chargées correctement
if img1 is None or img2 is None:
    print("Erreur : Impossible de charger les images.")
    exit()

# Initialiser SIFT
sift = cv2.SIFT_create()

# Détecter les points clés et calculer les descripteurs
keypoints1, des1 = sift.detectAndCompute(img1, None)
keypoints2, des2 = sift.detectAndCompute(img2, None)

# Vérifier si les descripteurs sont valides
if des1 is None or des2 is None:
    print("Erreur : Aucun descripteur trouvé dans l'une des images.")
    exit()

# Créer un objet FLANN
index_params = dict(algorithm=1, trees=5)  # Utiliser l'algorithme KDTREE
search_params = dict(checks=50)  # Nombre de vérifications pour chaque arbre

flann = cv2.FlannBasedMatcher(index_params, search_params)

# Matcher les descripteurs
matches = flann.knnMatch(des1, des2, k=2)

# Appliquer le ratio test de Lowe
good_matches = []
for m, n in matches:
    if m.distance < 0.7 * n.distance:  # Seuil de 0.7
        good_matches.append(m)

# Extraire les points correspondants
src_pts = np.float32([keypoints1[m.queryIdx].pt for m in good_matches]).reshape(-1, 1, 2)
dst_pts = np.float32([keypoints2[m.trainIdx].pt for m in good_matches]).reshape(-1, 1, 2)

# Appliquer RANSAC pour éliminer les correspondances aberrantes
if len(good_matches) > 7:  # Au moins 8 correspondances nécessaires pour RANSAC
    M, mask = cv2.findHomography(src_pts, dst_pts, cv2.RANSAC, 5.0)
    matches_mask = mask.ravel().tolist()
else:
    print("Pas assez de bonnes correspondances pour appliquer RANSAC.")
    matches_mask = None

# Dessiner les meilleures correspondances
draw_params = dict(matchColor=(0, 255, 0), singlePointColor=(255, 0, 0), matchesMask=matches_mask, flags=0)
img_matches = cv2.drawMatches(img1, keypoints1, img2, keypoints2, good_matches, None, **draw_params)

# Afficher le résultat
plt.imshow(img_matches)
plt.axis('off')  # Ne pas afficher les axes
plt.show()
```
