```python
import cv2
import numpy as np
import os

# Function to create a checkerboard image
def create_checkerboard_image(rows, cols, square_size, filename):
    # Create a black image
    checkerboard = np.zeros((rows * square_size, cols * square_size), dtype=np.uint8)

    # Fill the checkerboard pattern
    for i in range(rows):
        for j in range(cols):
            if (i + j) % 2 == 0:
                checkerboard[i * square_size:(i + 1) * square_size, j * square_size:(j + 1) * square_size] = 255

    # Save the image
    cv2.imwrite(filename, checkerboard)

# Parameters for the checkerboard
rows = 6  # Number of rows of squares
cols = 8  # Number of columns of squares
square_size = 50  # Size of each square in pixels

# Create a directory to save images
output_dir = 'checkerboard_images'
os.makedirs(output_dir, exist_ok=True)

# Generate and save 3 checkerboard images
for i in range(3):
    filename = os.path.join(output_dir, f'checkerboard_{i + 1}.png')
    create_checkerboard_image(rows, cols, square_size, filename)
    print(f'Checkerboard image saved as: {filename}')
```

```python
import cv2
import numpy as np
import os
from google.colab.patches import cv2_imshow  # Pour afficher les images dans Google Colab

# Définir le nombre de coins internes dans le motif d'échiquier
chessboard_size = (7, 5)  # (colonnes, lignes) - pour un échiquier 8x6
square_size = 50  # Taille des carrés en pixels

# Préparer les points d'objet
objp = np.zeros((chessboard_size[0] * chessboard_size[1], 3), np.float32)
objp[:, :2] = np.mgrid[0:chessboard_size[0], 0:chessboard_size[1]].T.reshape(-1, 2) * square_size

# Listes pour stocker les points d'objet et les points d'image
objpoints = []  # Points 3D dans le monde réel
imgpoints = []  # Points 2D dans l'image

# Charger les images d'échiquier
image_filenames = [os.path.join('checkerboard_images', f'checkerboard_{i + 1}.png') for i in range(3)]

for fname in image_filenames:
    img = cv2.imread(fname)
    if img is None:
        print(f"Erreur de chargement de l'image : {fname}")
        continue

    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    gray = cv2.equalizeHist(gray)  # Égalisation de l'histogramme pour améliorer le contraste

    # Trouver les coins de l'échiquier
    ret, corners = cv2.findChessboardCorners(gray, chessboard_size,
                                             cv2.CALIB_CB_ADAPTIVE_THRESH |
                                             cv2.CALIB_CB_NORMALIZE_IMAGE)

    # Si trouvé, ajouter les points d'objet et d'image
    if ret:
        objpoints.append(objp)
        imgpoints.append(corners)
        cv2.drawChessboardCorners(img, chessboard_size, corners, ret)
        cv2_imshow(img)  # Afficher l'image avec les coins détectés
        cv2.waitKey(500)  # Attendre 500 ms pour voir l'image
    else:
        print(f"Aucun coin d'échiquier détecté dans l'image : {fname}")

# Calibration de la caméra
if len(objpoints) > 0:  # Assurez-vous qu'il y a des points pour la calibration
    h, w = gray.shape[:2]  # Obtenir la taille de l'image
    ret, mtx, dist, rvecs, tvecs = cv2.calibrateCamera(objpoints, imgpoints, (w, h), None, None)

    # Afficher les résultats
    print("Matrice de la caméra :")
    print(mtx)
    print("Coefficients de distorsion :")
    print(dist)
    print("Vecteurs de rotation :")
    print(rvecs)
    print("Vecteurs de translation :")
    print(tvecs)
else:
    print("Aucun coin d'échiquier détecté dans les images fournies.")
```
